```python
# Carregando os datasets com os caminhos absolutos
netflix_data = pd.read_csv(netflix_path)
hulu_data = pd.read_csv(hulu_path)
amazon_data = pd.read_csv(amazon_path)

# Exibindo as primeiras linhas de cada dataset
print("Netflix:")
print(netflix_data.head())
print("\nHulu:")
print(hulu_data.head())
print("\nAmazon:")
print(amazon_data.head())
```

```python
# Verificando dados faltantes
print("Netflix Missing Data:")
print(netflix_data.isnull().sum())

print("\nHulu Missing Data:")
print(hulu_data.isnull().sum())

print("\nAmazon Missing Data:")
print(amazon_data.isnull().sum())

# Preenchendo valores faltantes
netflix_data = netflix_data.fillna('Unknown')
hulu_data = hulu_data.fillna('Unknown')
amazon_data = amazon_data.fillna('Unknown')
```
